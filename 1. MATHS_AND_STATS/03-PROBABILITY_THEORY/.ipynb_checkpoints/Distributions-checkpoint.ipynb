```python
import torch
```

```python
# १. डिस्ट्रीब्यूशन बनाना (Distribution)
# मान लीजिए मॉडल का 'औसत' (Mean) है 20 डिग्री
# और 'बिखराव' (Standard Deviation) है 5
# (अगर यह 5 के बजाय 0.1 होता, तो मॉडल बहुत 'Sure' होता)
mean = torch.tensor([20.0])
std_dev = torch.tensor([5.0])

dist = torch.distributions.Normal(mean, std_dev)
```

```python
# २. सैंपलिंग (Sampling)
# मॉडल इस 'पैटर्न' में से एक अंदाज़ा लगाता है
prediction = dist.sample()

print(f"Model ka andaja (Prediction): {prediction.item():.2f}")
```

```python
# ३. लॉग प्रोबेबिलिटी (Log Probability) - एआई इंजीनियरिंग का असली हिस्सा
# यह हमें बताता है कि यह भविष्यवाणी कितनी 'संभावित' है
# अगर वैल्यू बहुत कम है, तो मतलब यह 'अनिश्चित' (Uncertain) है
log_prob = dist.log_prob(prediction)
print(f"Iss bhavisyavani ka confidence (Log Prob): {log_prob.item():.4f}")
```
