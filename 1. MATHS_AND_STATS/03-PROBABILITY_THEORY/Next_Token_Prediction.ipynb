```python
import torch
import torch.nn.functional as F
```

```python
# Agle 5 sambhavit shabdon ke scores (Logits)
logits = torch.tensor([10.5, 10.0, 8.4, 1.1, -2.0, 5.0]) 
```

```python
# Kam temperature se confidence badh jaata hai
T_low = 0.5

# Jada temperature se creativity aur hallucination badh jaata hai
T_high = 2.0
```

```python
# Softmax se Low Temperature Probability nikalna
probs_low = F.softmax(logits / T_low, dim=0)
print(f"Probabilities: {probs_low}")
```

```python
# Softmax se High Temperature Probability nikalna
probs_high = F.softmax(logits / T_high, dim=0)
print(f"Probabilities: {probs_high}")
```

```python
# Multinomial Sampling: Probability ke hisaab se ek shabd chun-na
# Jiski probability zyada hai, uske chune jaane ke chance zyada hain
next_token_index = torch.multinomial(probs_low, num_samples=1)

print(f"Chuna gaya shabd (Index): {next_token_index.item()}")
```

```python
# Multinomial Sampling: Probability ke hisaab se ek shabd chun-na
# Jiski probability zyada hai, uske chune jaane ke chance zyada hain
next_token_index = torch.multinomial(probs_high, num_samples=1)

print(f"Chuna gaya shabd (Index): {next_token_index.item()}")
```
