```python
import torch
```

```python
raw_data = torch.tensor([10.2, 11.5, 9.8, 12.1, 10.5, 5000.0])
```

```python
mean_val = torch.mean(raw_data)
```

```python
median_val = torch.median(raw_data)
```

```python
variance_val = torch.var(raw_data)
```

```python
std_val = torch.std(raw_data)
```

```python
print("Mean=", mean_val.item())
print("Median=", median_val.item())
print("Variance=", variance_val.item())
print("Standard Deviation=", std_val.item())
```
