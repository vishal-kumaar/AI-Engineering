```python
import torch
```

```python
# 1. Probability distribution (High certainty: 90% chance of one class)
probs_low_entropy = torch.tensor([0.9, 0.1])
```

```python
# 2. Probability distribution (High uncertainty: 50/50 chance)
probs_high_entropy = torch.tensor([0.5, 0.5])
```

```python
def calculate_entropy(p):
    # Standard formula: -sum[p(x) * log2p(x)]
    return -torch.sum(p * torch.log2(p))
```

```python
print(f"Low Entropy (Sure): {calculate_entropy(probs_low_entropy):.2f}")
print(f"High Entropy (Confused): {calculate_entropy(probs_high_entropy):.2f}")
```
