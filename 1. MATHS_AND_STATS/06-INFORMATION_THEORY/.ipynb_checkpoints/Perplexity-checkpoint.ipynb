```python
# High Perplexity: It means AI model will have multiple options with identical probablities.
# Low Perplexity: It means AI model will have multiple options with different probabilities.
```

```python
import torch
```

```python
# 1. Suppose we have the average Cross-Entropy loss of an LLM
avg_loss = torch.tensor(2.3) 
```

```python
# 2. Perplexity calculation: exp(loss)
# Logic: It converts logarithmic bits back into number of choices
perplexity = torch.exp(avg_loss)
```

```python
print(f"Cross-Entropy Loss: {avg_loss.item():.2f}")
print(f"Perplexity (Avg. branching factor): {perplexity.item():.2f}")
```
