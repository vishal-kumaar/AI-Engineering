```python
import torch
import torch.nn as nn
```

```python
# 1. Actual Label (The Truth)
# Suppose index 0 is 'Cat', index 1 is 'Dog'
target = torch.tensor([0]) # Truth is 'Cat'
```

```python
# 2. Model's Predicted Scores (Logits)
# High score for Cat, Low for Dog
logits_good = torch.tensor([[2.5, -1.0]])

# Low score for Cat, High for Dog (Wrong prediction)
logits_bad = torch.tensor([[-0.5, 3.0]])
```

```python
# 3. Using PyTorch CrossEntropyLoss
criterion = nn.CrossEntropyLoss()

loss_good = criterion(logits_good, target)
loss_bad = criterion(logits_bad, target)
```

```python
print(f"Loss when model is correct: {loss_good.item():.4f}")
print(f"Loss when model is wrong: {loss_bad.item():.4f}")
```
