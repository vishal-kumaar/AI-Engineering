```python
import torch
```

```python
# 1. Taiyari (Setup)
x = torch.tensor(2.0)
target = torch.tensor(10.0)
w = torch.tensor(1.0, requires_grad=True) # Galat shuruat (1.0 * 2.0 = 2.0, humein 10 chahiye)

learning_rate = 0.1 # Chote kadam

print(f"Shuruati Weight: {w.item()}")
```

```python
# 2. Training Loop (Model seekh raha hai)
for epoch in range(20):
    # Forward Pass: Prediction nikalna
    y_pred = x * w
    
    # Loss nikalna: (Target - Prediction)^2
    loss = (target - y_pred)**2
    
    # Backward Pass: Galti ki sensitivity (Partial Derivative) nikalna
    loss.backward()
    
    # Gradient Descent: Weight ko update karna
    # Hum w.data use karte hain taaki update process khud gradient na banne lage
    with torch.no_grad():
        w -= learning_rate * w.grad
        
    # Agli baar ke liye gradient ko zero karna zaroori hai
    w.grad.zero_()
    
    if (epoch + 1) % 5 == 0:
        print(f"Epoch {epoch+1}: Loss = {loss.item():.4f}, Weight = {w.item():.4f}")
```

```python
print(f"\nTraining ke baad Weight: {w.item():.4f}")
print(f"Final Prediction: {x.item() * w.item():.4f}")
```
