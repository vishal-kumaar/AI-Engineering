```python
import torch
import torch.nn.functional as F
```

```python
# 1. Model ke raw outputs (Logits)
# Maano: [Billi ka score, Kutte ka score, Sher ka score]
logits = torch.tensor([2.0, 1.0, 0.1])
```

```python
# 2. Inhe Probability mein badalna (Softmax ka istemal)
# Softmax sabko 0 aur 1 ke beech le aayega aur jod (sum) 1 kar dega
probabilities = F.softmax(logits, dim=0)
```

```python
print(f"Probabilities: {probabilities}")
```
