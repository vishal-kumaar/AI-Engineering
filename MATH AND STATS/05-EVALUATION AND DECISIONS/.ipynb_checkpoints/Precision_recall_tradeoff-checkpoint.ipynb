```python
import torch
```

```python
# 1. ACTUAL values (1 = Fraud, 0 = Normal)
actual = torch.tensor([1, 0, 1, 0, 0, 1, 0])
```

```python
# 2. MODEL PREDICTIONS
predicted = torch.tensor([0, 1, 1, 1, 0, 1, 0])
```

```python
# LOGIC for Mistakes:
# False Positive (FP): Predicted 1, but Actual was 0
false_positives = torch.sum((predicted == 1) & (actual == 0))

# False Negative (FN): Predicted 0, but Actual was 1
false_negatives = torch.sum((predicted == 0) & (actual == 1))

# True Positive (TN): Predicted 1 and Actual was 1
true_positives = torch.sum((predicted == 1) & (actual == 1))
```

```python
print(f"False Positives: {false_positives.item()}") # Model ne galti se Fraud bola
print(f"False Negatives: {false_negatives.item()}") # Model ne Fraud ko miss kiya
print(f"True Positive: {true_positives.item()}") # Model ne sahi Fraud pakda
```

```python
precision = (true_positives) / (true_positives + false_positives)
recall = (true_positives) / (true_positives + false_negatives)
```

```python
print(f"Precision: {precision.item():.2f}")
print(f"Recall: {recall.item():.2f}")
```

```python
# - When we increase precision than it automatically decrease recall and 
# - When we increase recall than it automatically decrease precision.
# This is Precision vs Recall Tradeoff
```

```python
# Threshold Tuning
# AI don't give "yes" or "no" in response, but they give probability such as "0.85" means "85%".
# Default threshold: If probability greater than 0.5 then "yes", otherwise "no".
# For better precision: We set higher threshold such as "0.9". For example, in spam filter in email.
# For better recall: We set lower threshold such as "0.1". For example, in security alarm.
```
