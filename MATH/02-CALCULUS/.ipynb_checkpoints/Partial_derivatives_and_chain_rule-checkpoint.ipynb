```python
import torch
```

```python
# 1. Input aur Weights (requires_grad=True taaki PyTorch inka gradient yaad rakhe)
x = torch.tensor(2.0)
w1 = torch.tensor(3.0, requires_grad=True)
w2 = torch.tensor(4.0, requires_grad=True)
```

```python
# 2. Forward Pass (Kadi se kadi jodna)
# Layer 1 ka output (h) depend karta hai w1 par
h = x * w1 

# Final Output (y) depend karta hai h par (isliye w1 par bhi)
y = h * w2
```

```python
# 3. Maan lijiye humein y = 25 chahiye tha, par aaya kuch aur
target = torch.tensor(25.0)

# Loss function: y - target
loss = y - target
```

```python
# 4. BACKWARD PASS (Yahan Chain Rule chalta hai)
loss.backward()
```

```python
# 5. Ab dekhte hain w1 ka Partial Derivative
# PyTorch ne piche se aakar (Loss -> y -> h -> w1) sabko multiply kar diya hai
print(f"Loss ka partial derivative w1 ke liye: {w1.grad}")
```

```python
# 5. Ab dekhte hain w2 ka Partial Derivative
# PyTorch ne piche se aakar (Loss -> y -> w2) sabko multiply kar diya hai
print(f"Loss ka partial derivative w1 ke liye: {w2.grad}")
```

```python
# Topic: Chain Rule aur Backpropagation ka Jadoo

# Goal:
# Hame ye janna hai ki w1 ko badalne se Loss (L) par kitna asar padega?
# Maths ki bhasha mein: dL/dw1

# ---

# Kadi 1: L par y ka kya asar hai?
# Equation: L = y - 25
# Derivative (dL/dy) = 1
# (Matlab: Agar y ek kadam badhega, toh L bhi ek kadam badhega)

# Kadi 2: y par h ka kya asar hai?
# Equation: y = h * w2 (Jahan w2 = 4)
# Derivative (dy/dh) = w2 = 4
# (Matlab: Agar h ek kadam badhega, toh y char kadam badhega)

# Kadi 3: h par w1 ka kya asar hai?
# Equation: h = x * w1 (Jahan x = 2)
# Derivative (dh/dw1) = x = 2
# (Matlab: Agar w1 ek kadam badhega, toh h do kadam badhega)

# ---

# Aakhiri Hisab (The Multiplication):
# Ab in teeno kadiyon ko aapas mein multiply kar dijiye:

# dL/dw1 = (dL/dy) * (dy/dh) * (dh/dw1)
# dL/dw1 = 1 * 4 * 2 = 8

# ---

# Common Sense Result:
# Is 8 ka matlab ye hai ki agar aap w1 ko sirf 1 unit badhayenge, toh aapka final result (Loss) 8 unit badal jayega.
```

```python
# Topic: Chain Rule (w2 ke liye)

# Goal: 
# Hame ye janna hai ki w2 ko badalne se Loss (L) par kitna asar padega?
# Maths ki bhasha mein: dL/dw2

# ---

# Kadi 1: Loss par y ka asar
# Equation: L = y - 25
# Derivative (dL/dy) = 1
# (Matlab: Agar y ek kadam badhega, toh L bhi ek kadam badhega)

# Kadi 2: y par w2 ka asar
# Equation: y = h * w2
# Derivative (dy/dw2) = h
# (Pichle example ke hisab se h = 6 tha)

# ---

# Aakhiri Hisab (The Multiplication for w2):
# Ab in dono kadiyon ko aapas mein multiply kar dijiye:

# dL/dw2 = (dL/dy) * (dy/dw2)
# dL/dw2 = 1 * 6 = 6

# ---

# Conclusion:
# Iska matlab ye hai ki agar aap w2 ko 1 unit badhayenge, toh Loss 6 unit badh jayega.
```
