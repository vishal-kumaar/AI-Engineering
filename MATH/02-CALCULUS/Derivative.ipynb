```python
import torch
```

```python
# 1. Target (Sahi Jawab) aur Prediction (Model ka jawab)
target = torch.tensor([10.0])
prediction = torch.tensor([8.0], requires_grad=True) # AI ka jawab
```

```python
# 2. Loss Function (Humne decide kiya: MSE)
loss = (target - prediction)**2 # (10 - 8)^2 = 4
```

```python
# 3. Calculus (Backward) - Ye AI ko batata hai ki kitna badalna hai
loss.backward()
```

```python
# Ye hai wo sensitivity (Derivative)
print(f"Loss ki sensitivity prediction ke muqable: {prediction.grad}")
```

```python
# 1. Weight define kiya (Maano humara purana 2.0 wala weight)
w = torch.tensor([3.0], requires_grad=True)
```

```python
# 2. Function: Loss = w^2
loss = w**2
```

```python
# 3. Ye step "Hota kaise hai" ka engine hai
loss.backward() # Ye piche jaakar har weight ka derivative nikaal deta hai
```

```python
# 4. Check karo slope (derivative) kitni aayi
print(f"Weight ki slope (Derivative): {w.grad}") # Output: 6.0
```
