```python
import torch
```

```python
# 1. Prior: P(Spam) - Pehle se pata probability
p_spam = torch.tensor(0.20)
```

```python
# 2. Likelihood: P(Free | Spam) - Spam mein 'Free' hone ki sambhavna
p_free_given_spam = torch.tensor(0.90)
```

```python
# 3. Evidence: P(Free) - Kisi bhi email mein 'Free' hone ki probability
p_free = torch.tensor(0.30)
```

```python
# 4. Bayes Theorem Formula: P(A|B) = [P(B|A) * P(A)] / P(B)
# Yahan: P(Spam|Free) = [P(Free|Spam) * P(Spam)] / P(Free)
p_spam_given_free = (p_free_given_spam * p_spam) / p_free
```

```python
print(f"Pehle sambhavna thi: {p_spam.item()*100:.2f}%")
print(f"'Free' shabd dekhne ke baad nayi sambhavna: {p_spam_given_free.item()*100:.2f}%")
```
