```python
import torch
import matplotlib.pyplot as plt
```

```python
# 1. NORMAL DISTRIBUTION (Symmetric)
# Mean = 0, Std Dev = 1
normal_data = torch.tensor([5, 7, 8, 9, 10, 10, 10, 10, 11, 12, 13, 15], dtype=torch.float)
```

```python
# 2. POSITIVE SKEW (Right Skewed)
# Using Exponential distribution to create a tail on the right
positive_skew = torch.tensor([1, 2, 2, 3, 3, 3, 4, 5, 10, 50, 100], dtype=torch.float)
```

```python
# 3. NEGATIVE SKEW (Left Skewed)
# Flipping the positive skew to create a tail on the left
negative_skew = torch.tensor([10, 40, 85, 90, 95, 96, 97, 98, 99, 100], dtype=torch.float)
```

```python
# VISUALIZATION
plt.figure(figsize=(15, 5))
```

```python
# Plotting Normal Distribution
plt.subplot(1, 3, 1)
plt.hist(normal_data.numpy(), bins=30, color='blue', alpha=0.6)
plt.title('Normal Distribution')

plt.tight_layout()
```

```python
# Plotting Positive Skew
plt.subplot(1, 3, 2)
plt.hist(positive_skew.numpy(), bins=30, color='green', alpha=0.6)
plt.title('Positive Skew (Right)')

plt.tight_layout()
```

```python
# Plotting Negative Skew
plt.subplot(1, 3, 3)
plt.hist(negative_skew.numpy(), bins=30, color='red', alpha=0.6)
plt.title('Negative Skew (Left)')

plt.tight_layout()
```
