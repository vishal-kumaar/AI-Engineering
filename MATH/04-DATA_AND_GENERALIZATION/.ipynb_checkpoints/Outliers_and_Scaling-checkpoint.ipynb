```python
import torch
```

```python
# 1. Manual Data with an extreme outlier (5000.0)
data = torch.tensor([10.2, 12.5, 11.0, 9.5, 10.8, 5000.0])
```

```python
# 2. Logic to find Outliers using Z-Score
mean = data.mean()
std = data.std()
z_scores = (data - mean) / std
```

```python
# Identifying points where Z-Score > 2 (Standard threshold)
threshold = 2
is_outlier = torch.abs(z_scores) > threshold
clean_data = data[~is_outlier] # ~ means 'NOT'
```

```python
print(f"Original Data: {data}")
print(f"Clean Data: {clean_data}")
```

```python
# A. Standardization (Mean=0, Std=1)
# Formula: (x - mean) / std
standardized = (clean_data - clean_data.mean()) / clean_data.std()
```

```python
# B. Min-Max Scaling (Range 0 to 1)
# Formula: (x - min) / (max - min)
min_val = clean_data.min()
max_val = clean_data.max()
normalized = (clean_data - min_val) / (max_val - min_val)
```

```python
print(f"Standardized: {standardized}")
print(f"Min-Max Normalized: {normalized}")
```
