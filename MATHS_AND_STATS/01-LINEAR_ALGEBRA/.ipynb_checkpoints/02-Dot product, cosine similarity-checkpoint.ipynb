```python
import torch
import torch.nn.functional as F
```

```python
v1 = torch.tensor([1.0, 2.0, 3.0])
v2 = torch.tensor([4.0, 5.0, 6.0])
```

```python
dot_prod = torch.dot(v1, v2)

print(dot_prod)
```

```python
manual_dot_prod = torch.sum(v1 * v2)

print(manual_dot_prod)
```

```python
#              (v1 . v2)
# cosine_sim = ---------
#              \\v1\\v2\\

sim = F.cosine_similarity(v1.unsqueeze(0), v2.unsqueeze(0))
print("Cosine Similarity:", sim.item())
```

```python
# Cosine similarity manually
cosine_sim = torch.dot(v1, v2) / (torch.norm(v1) * torch.norm(v2))

print(cosine_sim)
```

```python
# Angle between two vectors
angle_rad = torch.acos(cosine_sim)
angle_deg = torch.rad2deg(angle_rad)
print(angle_deg)
```
