```python
import torch
```

```python
# 1. Ek simple function banate hain: f(x) = x^2 + 5
def my_ai_function(x):
    return x**2 + 5
```

```python
# 2. Input dete hain (Tensor)
x_input = torch.tensor(3.0)
```

```python
# 3. Output nikalte hain
y_output = my_ai_function(x_input)
```

```python
print(f"Input: {x_input.item()}, Output: {y_output.item()}") 
```
