```python
import torch
```

```python
# 1. Suppose we have model accuracy scores from 10 different tests
scores = torch.tensor([0.82, 0.85, 0.84, 0.88, 0.86, 0.83, 0.85, 0.87, 0.84, 0.85])
```

```python
# 2. Calculating Mean and Standard Error
mean_accuracy = scores.mean()
std_error = scores.std() / torch.sqrt(torch.tensor(len(scores)))
```

```python
# 3. 95% Confidence Interval (Using 1.96 as standard Z-score for 95%)
lower_bound = mean_accuracy - (1.96 * std_error)
upper_bound = mean_accuracy + (1.96 * std_error)
```

```python
print(f"Model Accuracy: {mean_accuracy:.2f}")
print(f"95% Confidence Interval: [{lower_bound:.2f} to {upper_bound:.2f}]")
```
