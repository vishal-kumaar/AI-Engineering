```python
# KL Divergence: It tells the information loss between two distribution
```

```python
import torch
import torch.nn.functional as F
```

```python
# Two probability distributions
P = torch.tensor([0.4, 0.6]) # Target
Q = torch.tensor([0.3, 0.7]) # Model approximation
```

```python
# KL Divergence logic: P * log(P/Q)
kl_div = F.kl_div(Q.log(), P, reduction='batchmean')

print(f"KL Divergence (Difference between P and Q): {kl_div.item():.4f}")
```
