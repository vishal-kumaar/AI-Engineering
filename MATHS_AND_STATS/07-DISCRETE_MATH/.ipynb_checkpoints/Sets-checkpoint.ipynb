```python
import torch
```

```python
# 1. Manual Sets of User IDs
set_a = torch.tensor([101, 102, 103, 104, 105]) # Users who clicked
set_b = torch.tensor([104, 105, 106, 107])      # Users who purchased
```

```python
# 2. Operations (Logic)
# Intersection: Users who clicked AND purchased
# We use numpy-like logic in PyTorch or specialized functions
combined = torch.cat((set_a, set_b))
uniques, counts = combined.unique(return_counts=True)
intersection = uniques[counts > 1]
```

```python
print(f"Users who did both: {intersection}")
```
