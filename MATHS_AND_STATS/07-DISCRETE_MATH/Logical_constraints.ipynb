```python
def agent_decision(is_logged_in, has_subscription):
    # Logical AND constraint
    can_access = is_logged_in and has_subscription
    
    if can_access:
        return "Access Granted"
    else:
        return "Access Denied"
```

```python
# Manual Testing
print(f"User 1: {agent_decision(True, False)}")
print(f"User 2: {agent_decision(True, True)}")
```
